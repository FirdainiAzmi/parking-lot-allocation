# Scrap Gizi

```python
import requests as rs
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#WAJIB DI RUN, TAPI SEKALI AJA (buat ngejaga dataframe)
v = 0
```

```python
root = "https://www.fatsecret.co.id/kalori-gizi/umum/daging-sapi-panggang"

nama = []
berat = []

kalori = []
karbo = []
protein = []
lemak = []

energi = []
lemak_jenuh = []
kolestrol = []
serat = []
gula = []
sodium = []
kalium = []

html = rs.get(root)
soup = BeautifulSoup(html.text, features="html.parser")
info_nama = soup.find_all('div', class_='summarypanelcontent')

for x in info_nama :
    nama.append((x.find('h1')).text)

info = soup.find('div', class_='nutrition_facts international')
list1 = info.find_all('div', class_='nutrient black right tRight')

jumlah_energi = list1[0].text.strip()
energi.append(jumlah_energi)

jumlah_lemak = list1[1].text.strip()
lemak.append(jumlah_lemak)

jumlah_protein = list1[2].text.strip()
protein.append(jumlah_protein)

jumlah_karbo = list1[3].text.strip()
karbo.append(jumlah_karbo)

list2 = info.find_all('div', class_='nutrient right tRight')

jumlah_kalori = list2[0].text.strip()
kalori.append(jumlah_kalori)

jumlah_lemak_jenuh = list2[1].text.strip()
lemak_jenuh.append(jumlah_lemak_jenuh)

jumlah_kolestrol = list2[4].text.strip()
kolestrol.append(jumlah_kolestrol)

jumlah_serat = list2[5].text.strip()
serat.append(jumlah_serat)
# serat.append("-")

jumlah_gula = list2[6].text.strip()
gula.append(jumlah_gula)
# gula.append("-")

jumlah_sodium = list2[7].text.strip()
sodium.append(jumlah_sodium)

jumlah_kalium = list2[8].text.strip()
kalium.append(jumlah_kalium)

list3 = info.find_all('div', class_='serving_size black serving_size_value')

jumlah_berat = list3[0].text.strip()
berat.append(jumlah_berat)

if v == 0 :
  df = pd.DataFrame({
      'Nama' : nama,
      'Karbohidrat' : karbo,
      'Protein' : protein,
      'kalori' : kalori,
      'Lemak' : lemak,
      'Energi' : energi,
      'lemak jenuh' : lemak_jenuh,
      'kolestrol' : kolestrol,
      'serat' : serat,
      'gula' : gula,
      'sodium' : sodium,
      'kalium' : kalium,
      'berat' : berat
  })
  v += 1
else :
  df_new = pd.DataFrame({
      'Nama' : nama,
      'Karbohidrat' : karbo,
      'Protein' : protein,
      'kalori' : kalori,
      'Lemak' : lemak,
      'Energi' : energi,
      'lemak jenuh' : lemak_jenuh,
      'kolestrol' : kolestrol,
      'serat' : serat,
      'gula' : gula,
      'sodium' : sodium,
      'kalium' : kalium,
      'berat' : berat
  })
  df = pd.concat([df, df_new], ignore_index=True)

df.head(100)
```

# Image Scrap

```python
import requests
import os
import time

# Ganti ini dengan ACCESS KEY dari aplikasi Unsplash milikmu
ACCESS_KEY = ''

# Kata kunci pencarian
query = 'white rice'
jumlah_total = 50
folder = 'nasi_putih'

os.makedirs(folder, exist_ok=True)

gambar_terkumpul = 0
halaman = 1
gambar_per_halaman = 30  # batas dari Unsplash

while gambar_terkumpul < jumlah_total:
    sisa = jumlah_total - gambar_terkumpul
    ambil = min(sisa, gambar_per_halaman)

    url = f"https://api.unsplash.com/search/photos?query={query}&per_page={ambil}&page={halaman}&client_id={ACCESS_KEY}"
    response = requests.get(url)
    data = response.json()
    hasil = data.get('results', [])

    if not hasil:
        print(" Tidak ada lagi gambar yang tersedia.")
        break

    for i, item in enumerate(hasil):
        image_url = item['urls']['regular']
        image_data = requests.get(image_url).content
        filename = f"{folder}/{query.replace(' ', '_')}_{gambar_terkumpul + 1}.jpg"

        with open(filename, 'wb') as f:
            f.write(image_data)
        print(f" Gambar {gambar_terkumpul + 1} disimpan")

        gambar_terkumpul += 1
        if gambar_terkumpul >= jumlah_total:
            break

    halaman += 1
    time.sleep(1)  # delay kecil untuk menghindari rate limit

print(f"\n Selesai mengunduh {gambar_terkumpul} gambar dari '{query}'")
```
